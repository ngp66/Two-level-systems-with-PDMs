## Time evolution

Let us define parameters

```python
# all parameters are in atomic units
hbar = 1
w0 = 5  # atom transition frequency 
wL = 5  # laser frequency
det = wL - w0  # detuning from resonance
d = 1  # transition dipole moment
E = 0.2  # electric field amplitude
y = 0.08 # spontaneous emission coefficient
times = np.linspace(0, 100, 10001)  # creates list of 10001 points evenly distributed between values 0 and 100
```

Let us define basis

```python
e = basis(2, 0)  # excited state
g = basis(2, 1)  # ground state
```

```python
e
```

```python
g
```

Initial state as a density matrix

```python
rho0 = g * g.dag()  # initially system in the ground state
```

```python
rho0
```

Let us define Hamiltonian with separated time-independent and time-dependent parts. There is a special for for the time-depnedent part as shown below. Time should be expressed as `t` and there are available standard functions as `sin`, `cos`, `exp`, ... .

```python
H0 = 0.5 * hbar * w0 * sigmaz()  # time-independent part
H1 = [-0.5 * E * d * sigmax(), "exp(1j*wL*t) + exp(-1j*wL*t)"]  # time-dependent part as a list where second term is f(t)
H = [H0, H1]  # the whole Hamiltonian has to be write as a list of all parts
```

```python
H
```

Finding solution of $i\hbar\dot{\rho} = [H, \rho]$ using `mesolve`. All the unknown variables in time-dependant part of the Hamiltonian has to be put in the `arg` argument as a dictionary

```python
rho = mesolve(H, rho0, times, [np.sqrt(y)*sigmam()], [], args = {"wL": wL}, progress_bar=None)
```

Once we have a object `rho` which contains all the results $\rho(t)$, we should be able to extract the results. We can use `.states`, which returns list of density matrices in each time step [$\rho(t_0)$, $\rho(t_0 + \Delta t)$, $\rho(t_0 + 2\Delta t)$, ..., $\rho(t_{end})$]

```python
rho.states[0]
```

```python
rho.states[-1]
```

## Plotting the results

To plot, we will use another package `matplotlib`

```python
from matplotlib import pyplot as plt
```

Now we can plot, e.g., expected value of $\sigma_z$ operator

```python
fig, ax = plt.subplots(1, figsize=(12, 5))  # create figure and axes (one plot here)
ax.plot(times, expect(sigmaz(), rho.states))  # plotting expected value of the sigma_z operator
ax.set_xlabel('times')  # set label of x-axis
ax.set_ylabel(r'Tr($\rho\cdot\sigma_z$)')  # set label of y-axes (LateX formulas are used here)
ax.grid()  # plot grid
plt.show()  # show the results
```

## Rotating wave approximation

The above solution is an exact solution as we have put full Hamiltonian. Here, we can compare this result with the Hamiltonian after rotating-wave (secular) approximation

```python
H_RWA = -0.5 * hbar * det * sigmaz() - 0.5 * E * d * sigmax()
```

```python
H_RWA
```

```python
rho_RWA = mesolve(H_RWA, rho0, times, [np.sqrt(y)*sigmam()], [], progress_bar=None)
```

We can plot both results in the same figure

```python
fig, ax = plt.subplots(1, figsize=(12, 5)) 
ax.plot(times, expect(sigmaz(), rho.states), label='full')  # plotting full <sigma_z>
ax.plot(times, expect(sigmaz(), rho_RWA.states), label='RWA')  # plotting RWA <sigma_z>
ax.set_xlabel('times')  # set label of x-axis
ax.set_ylabel(r'Tr($\rho\cdot\sigma_z$)')  # set label of y-axes (LateX formulas are used here)
ax.grid()  # plot grid
plt.legend()
plt.show()  # show the results
```

We can look as well on the coherences

```python
fig, ax = plt.subplots(1, figsize=(12, 5)) 
ax.plot(times, np.imag(expect(sigmap(), rho.states)), label='full')  # plotting full Im(<sigma_plus>)
ax.plot(times, np.imag(expect(sigmap(), rho_RWA.states)), label='RWA')  # plotting RWA Im(<sigma_plus>)
ax.set_xlabel('times')  # set label of x-axis
ax.set_ylabel(r'Tr($\rho\cdot\sigma_z$)')  # set label of y-axes (LateX formulas are used here)
ax.grid()  # plot grid
plt.legend()
plt.show()  # show the results
```

So indeed, we have pictures without fast-oscillating terms, which for the population removes small oscillations and for the coherences shows only the envelope.

## Analytical solution (no spontaneous emission, y = 0)

We can compare the above results with our analytical resuls as well\
    $\rho_{ee}(t) = \frac{\Omega^2_R}{\Omega^2}\sin^2(\frac{1}{2}\Omega t),$\
    $\rho_{eg}(t) = \frac{\Omega_R}{\Omega^2}\sin(\frac{1}{2}\Omega t)\left(-i\Omega\cos(\frac{1}{2}\Omega t) - \delta \sin(\frac{1}{2}\Omega t)\right),$\
    where $\Omega_R = Ed / \hbar$, $\Omega = \sqrt{\Omega^2_R + \delta^2}$

```python
Om_R = E * d / hbar
Om = np.sqrt(Om_R**2 + det**2)
rho_ee = (Om_R/Om)**2 * np.sin(0.5 * Om * times)**2
rho_eg = (Om_R/Om**2) * np.sin(0.5 * Om * times) * (-1j * Om * np.cos(0.5 * Om * times) - det * np.sin(0.5 * Om * times))
```

```python
fig, ax = plt.subplots(1, figsize=(15, 8)) 
ax.plot(times, expect(sigmaz(), rho.states), linewidth=2, label='full')  # plotting full <sigma_z>
ax.plot(times, expect(sigmaz(), rho_RWA.states), linewidth=2, label='RWA')  # plotting RWA <sigma_z>
ax.plot(times, 2 * rho_ee - 1, '--', linewidth=2, label='analytical')  # plotting analytical <sigma_z>
ax.set_xlabel('times')  # set label of x-axis
ax.set_ylabel(r'Tr($\rho\cdot\sigma_z$)')  # set label of y-axes (LateX formulas are used here)
ax.grid()  # plot grid
plt.legend()
plt.show()  # show the results
```

```python
fig, ax = plt.subplots(1, figsize=(15, 8)) 
ax.plot(times, np.imag(expect(sigmap(), rho.states)), linewidth = 2, label='full')  # plotting full Im(<sigma_plus>)
ax.plot(times, np.imag(expect(sigmap(), rho_RWA.states)), linewidth = 2, label='RWA')  # plotting RWA Im(<sigma_plus>)
ax.plot(times, np.imag(rho_eg), '--', linewidth = 2, label='analytical')  # plotting RWA Im(<sigma_plus>)
ax.set_xlabel('times')  # set label of x-axis
ax.set_ylabel(r'Tr($\rho\cdot\sigma_z$)')  # set label of y-axes (LateX formulas are used here)
ax.grid()  # plot grid
plt.legend()
plt.show()  # show the results
```

## Analytical solution (no detunning, delta = 0)

We can compare the above results with our analytical resuls as well\
    $\rho_{ee}(t) = \frac{\Omega^2_R}{\Omega^2}\sin^2(\frac{1}{2}\Omega t),$\
    $\rho_{eg}(t) = \frac{\Omega_R}{\Omega^2}\sin(\frac{1}{2}\Omega t)\left(i\Omega\cos(\frac{1}{2}\Omega t) + i\gamma \frac{1}{2}\sin(\frac{1}{2}\Omega t)\right),$\
    where $\Omega_R = Ed / \hbar$, $\Omega = \sqrt{\Omega^2_R - \frac{1}{4}\gamma^2}$

```python
Om_R = E * d / hbar

g = np.sqrt(Om_R**2 - (y**2)/16) #gamma

k = ((Om_R**2)/2)/(2*(y**2/4) + (Om_R**2)) #prefactor

k1 = (1j*Om_R*y/2)/(2*(y**2/4) + (Om_R**2))

rho_ee = k * (1 - (np.cos(g*times) + 3 * y * np.sin(g*times) / (4 * g)) * np.exp(-(3/4) * y * times))

rho_eg = k1 * (1 - (np.cos(g*times) + ((y**2)/4 - Om_R**2) * np.sin(g*times) / (y * g)) * np.exp(-(3/4) * y * times))
```

```python
fig, ax = plt.subplots(1, figsize=(15, 8)) 
ax.plot(times, expect(sigmaz(), rho.states), linewidth=2, label='full')  # plotting full <sigma_z>
ax.plot(times, expect(sigmaz(), rho_RWA.states), linewidth=2, label='RWA')  # plotting RWA <sigma_z>
ax.plot(times, 2*rho_ee - 1, '--', linewidth=2, label='analytical')  # plotting analytical <sigma_z>
ax.set_xlabel('times')  # set label of x-axis
ax.set_ylabel(r'Tr($\rho\cdot\sigma_z$)')  # set label of y-axes (LateX formulas are used here)
ax.grid()  # plot grid
plt.legend()
plt.show()  # show the results
```

```python
fig, ax = plt.subplots(1, figsize=(15, 8)) 
ax.plot(times, np.imag(expect(sigmap(), rho.states)), linewidth = 2, label='full')  # plotting full Im(<sigma_plus>)
ax.plot(times, np.imag(expect(sigmap(), rho_RWA.states)), linewidth = 2, label='RWA')  # plotting RWA Im(<sigma_plus>)
ax.plot(times, np.imag(rho_eg), '--', linewidth = 2, label='analytical')  # plotting RWA Im(<sigma_plus>)
ax.set_xlabel('times')  # set label of x-axis
ax.set_ylabel(r'Tr($\rho\cdot\sigma_z$)')  # set label of y-axes (LateX formulas are used here)
ax.grid()  # plot grid
plt.legend()
plt.show()  # show the results
```
